## 4. 모델링

### 4.1 원본 데이터로

```python
# Modeling with target of 'emotionPositive' using entire features.
modeling(df, 'emotionPositive', 'emotionTension', 'df')

# Modeling with target of 'emotionTension' using entire features.
modeling(df, 'emotionTension', 'emotionPositive', 'df')
```

### 4.2 센서 데이터만

```python
# Dataframe with sensor features only.
df_sensor = df[['e4Eda', 'e4Hr', 'e4Temp','emotionPositive', 'emotionTension']]

# Modeling with target of 'emotionPositive' using sensor features.
modeling(df_sensor, 'emotionPositive', 'emotionTension', 'df_sensor')

# Modeling with target of 'emotionTension' using sensor features.
modeling(df_sensor, 'emotionTension', 'emotionPositive', 'df_sensor')
```

## 5. XAI

```python
# Specifies the target data.
y_positive = df[['emotionPositive']]  # emotionPositive
y_tension = df[['emotionTension']]  # emotionTension

# Specifies the input data.
X = df.drop(['emotionTension', 'emotionPositive'], axis = 1)
X_train_positive, X_test_positive, y_train_positive, y_test_positive  = train_test_split(X,y_positive, test_size = 0.3, shuffle =True, random_state = 2, stratify=y_positive)
X_train_tension, X_test_tension, y_train_tension, y_test_tension  = train_test_split(X,y_tension, test_size = 0.3, shuffle =True, random_state = 2, stratify=y_tension)

# Initialize JavaScript visualization
shap.initjs()
```

### 5.1. emotionPositive

```python
shap_model('model_1_RandomForestClassifier_df_emotionPositive', X_train_positive, y_positive['emotionPositive'])
```

### 5.2. emotionTension

```python
shap_model('model_1_RandomForestClassifier_df_emotionTension', X_train_tension, y_tension['emotionTension'])
```
